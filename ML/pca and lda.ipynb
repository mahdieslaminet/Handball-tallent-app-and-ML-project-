1️⃣ ایمپورت‌ها و دیتاست

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

2️⃣ نرمال‌سازی

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

3️⃣ PCA با ۲ Component

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='coolwarm', alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA (2D Visualization)")
plt.colorbar(label="Class")
plt.show()
```

4️⃣ LDA با ۲ Component

```python
lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X_scaled, y)
```

```python
# Generate random jitter for better visualization of overlapping points in 1D
jitter_strength = 0.1
y_jittered = np.random.rand(len(X_lda)) * jitter_strength - jitter_strength / 2

plt.figure(figsize=(8, 4))
plt.scatter(X_lda[:, 0], y_jittered, c=y, cmap='coolwarm', alpha=0.7)
plt.xlabel("LDA Component 1")
plt.ylabel("") # No meaningful y-axis
plt.title("LDA (1D Visualization with Jitter)")
plt.yticks([]) # Hide y-axis ticks
plt.colorbar(label="Class")
plt.grid(axis='x', alpha=0.5)
plt.show()
```
